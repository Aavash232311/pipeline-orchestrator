```python
import pandas as pd

df = pd.read_csv("../Data/set/digitalSkillsCollection_en.csv")
print(len(df.columns.tolist()))
print(f"Preferred Labels: {df['preferredLabel'][162]}")
print(f"Alternate Labels: {df['altLabels'][162]}")
print(f"Broader Concepts: {df['broaderConceptPT'][162]}")
```

```python
df.columns.tolist()
```

### Meaningful data

```python
sub_frame = {
    "preferredLabel": df["preferredLabel"].fillna(''),
    "altLabels": df["altLabels"].fillna('') ,
    "broaderConceptPT": df["broaderConceptPT"].fillna('')
}


skill_df = pd.DataFrame(sub_frame)

labels = ["ASP.NET", "PHP", "Angular", "C++", "Javascript", "Visual Basic"]
labels_lower = [x.lower() for x in labels]
filtered_df = skill_df[skill_df['preferredLabel'].str.lower().isin(labels_lower)]
filtered_df  
```

## Let's try to extract programming languages here

<b>This has less useful information but I will consider it</b>

```python

concepts_lower = [c.lower() for c in ["computer programming"]]


programming_languages = skill_df[
    skill_df['broaderConceptPT']
    .str.split("|")
    .apply(lambda x: any(c.strip().lower() in concepts_lower for c in x) if isinstance(x, list) else False)
]

programming_languages_df = pd.DataFrame(programming_languages)
print(programming_languages_df.shape)
programming_languages_df
```
