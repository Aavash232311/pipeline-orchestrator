### Cosine Similarity

```python
import torch
from sentence_transformers import SentenceTransformer
import torch.nn.functional as F


str1 = "red"
str2 = "blue"

model = SentenceTransformer("sentence-transformers/all-MiniLM-L6-v2")

str1_emb = torch.from_numpy(model.encode(str1)).unsqueeze(0)
str2_emb = torch.from_numpy(model.encode(str2)).unsqueeze(0)
similarity = F.cosine_similarity(str1_emb, str2_emb, dim=1)
print(similarity)
```
